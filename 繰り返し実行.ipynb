```python
# ===== Driver Notebook: FS×Corr×正則化 全組合せ実行 =====
from itertools import product

NOTEBOOK_PATH = "実験データ解析_機械学習.ipynb"

FS_MODES = ["rank", "rfe"]
CORR_LIST = [1.0, 0.9, 0.8, 0.7]
REG_LEVELS = [0, 1]

for FS_MODE, CORR_THRESHOLD, XGB_REG_LEVEL in product(FS_MODES, CORR_LIST, REG_LEVELS):
    print("="*80)
    print(f"[RUN] FS_MODE={FS_MODE} | CORR_THRESHOLD={CORR_THRESHOLD} | XGB_REG_LEVEL={XGB_REG_LEVEL}")
    print("="*80)

    # ここで FS_MODE / CORR_THRESHOLD / XGB_REG_LEVEL が「グローバル変数」として
    # カーネル内に定義され、その状態で下の %run が実行される
    %run "$NOTEBOOK_PATH"
```
